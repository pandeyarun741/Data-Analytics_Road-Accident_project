```python

df['Fault of Driver-Total No. of Road Accidents - 2014'].fillna(df['Fault of Driver-Total No. of Road Accidents - 2014'].mean(), inplace=True)
df['Fault of Driver-Total No. of Road Accidents - 2014 per 1L people'].fillna(df['Fault of Driver-Total No. of Road Accidents - 2014 per 1L people'].mean(), inplace=True)

df['Other causes/causes not known-Number of Persons-Injured - 2014'].fillna(df['Other causes/causes not known-Number of Persons-Injured - 2014'].median(), inplace=True)


df.fillna(method='ffill', inplace=True)
```

```python
df['Total_No_Of_Road_Accident_Injured_Death'] = df[['Fault of Driver of other vehicles-Total No. of Road Accidents - 2014 per 1L people','Fault of Driver of other vehicles-Number of Persons-Injured - 2014 per 1L people','Fault of Driver of other vehicles-Number of Persons-Killed - 2014 per 1L people']].sum(axis=1)


total_No_Of_Road_Accident_Injured_Death= df[['States/UTs', 'Total_No_Of_Road_Accident_Injured_Death']].sort_values(by='Total_No_Of_Road_Accident_Injured_Death', ascending=False)
```

```python
df['Total_No_Of_Road_Accident_Injured_Death_by_Pedestrian'] = df[['Fault of Pedestrian-Total No. of Road Accidents - 2014 per 1L people','Fault of Pedestrian-Number of Persons-Killed - 2014 per 1L people','Fault of Pedestrian-Number of Persons-Injured - 2014 per 1L people']].sum(axis=1)


total_No_Of_Road_Accident_Injured_Death_by_Pedestrian= df[['States/UTs', 'Total_No_Of_Road_Accident_Injured_Death_by_Pedestrian']].sort_values(by='Total_No_Of_Road_Accident_Injured_Death_by_Pedestrian', ascending=False)

```

```python
df['Total_No_Of_Road_Accident_by_Pedestrian'] = df[['Fault of Pedestrian-Total No. of Road Accidents - 2014 per 1L people']].sum(axis=1)
df['Total_No_Of_Road_Injured_by_Pedestrian'] = df[['Fault of Pedestrian-Number of Persons-Killed - 2014 per 1L people']].sum(axis=1)
df['Total_No_Of_Road_Death_by_Pedestrian'] = df[['Fault of Pedestrian-Number of Persons-Injured - 2014 per 1L people']].sum(axis=1)


total_No_Of_Road_Accident_Injured_Death_by_Pedestrian = df[['States/UTs', 'Total_No_Of_Road_Accident_by_Pedestrian', 'Total_No_Of_Road_Injured_by_Pedestrian', 'Total_No_Of_Road_Death_by_Pedestrian']].sort_values(
    by=['Total_No_Of_Road_Accident_by_Pedestrian', 'Total_No_Of_Road_Injured_by_Pedestrian', 'Total_No_Of_Road_Death_by_Pedestrian'], 
    ascending=False
)


```
