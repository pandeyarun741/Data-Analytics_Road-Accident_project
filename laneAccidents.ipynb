```python
df['Total Accidents'] = df[['Single Lane - Accident - 2014 per 1L people', 
                            'Two Lanes - Accident - 2014 per 1L people', 
                            '3 Lanes or more w.o Median - Accident - 2014 per 1L people',
                            '4 Lanes with Median - Accident - 2014 per 1L people']].sum(axis=1)


state_wise_accidents = df[['State/UT', 'Total Accidents']].sort_values(by='Total Accidents', ascending=False)

```

```python
state_wise_accidents
```

```python
plt.figure(figsize=(12,8))
sns.barplot(y='Total Accidents', x='State/UT', data=state_wise_accidents, palette='viridis')

plt.title('State-wise Total Accidents for All Lane Types in 2014', fontsize=16)
plt.xlabel('State/UT', fontsize=14)
plt.ylabel('Total Accidents', fontsize=14)

# Show plot

plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
df['Total Injured'] = df[['Single Lane - Injured - 2014 per 1L people', 
                            'Two Lanes - Injured - 2014 per 1L people', 
                            '3 Lanes or more w.o Median - Injured - 2014 per 1L people',
                            '4 Lanes with Median - Injured - 2014 per 1L people']].sum(axis=1)


state_wise_injured= df[['State/UT', 'Total Injured']].sort_values(by='Total Injured', ascending=False)
```

```python
state_wise_injured
```

```python
plt.figure(figsize=(12,8))
sns.barplot(y='Total Injured', x='State/UT', data=state_wise_injured, palette='viridis')

plt.title('State-wise Total Injured for All Lane Types in 2014', fontsize=16)
plt.xlabel('State/UT', fontsize=14)
plt.ylabel('Total Injured', fontsize=14)

# Show plot
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
df['Total Killed'] = df[['Single Lane - Killed - 2014 per 1L people', 
                            'Two Lanes - Killed - 2014 per 1L people', 
                            '3 Lanes or more w.o Median - Killed - 2014 per 1L people',
                            '4 Lanes with Median - Killed - 2014 per 1L people']].sum(axis=1)


state_wise_Killed= df[['State/UT', 'Total Killed']].sort_values(by='Total Killed', ascending=False)
```

```python
df['Single line Total Accident, Killed, Injured'] = df[['Single Lane - Accident - 2014 per 1L people', 
                            'Single Lane - Killed - 2014 per 1L people', 
                            'Single Lane - Injured - 2014 per 1L people']].sum(axis=1)
                             

state_wise_Single_line_Accident_Killed_Injured= df[['State/UT', 'Single line Total Accident, Killed, Injured']].sort_values(by='Single line Total Accident, Killed, Injured', ascending=False)
```
