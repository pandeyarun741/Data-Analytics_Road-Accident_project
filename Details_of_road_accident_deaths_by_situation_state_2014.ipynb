```python
offender_deaths = df[['Offenders (Driver/Pedestrian) Died_Male',
                      'Offenders (Driver/Pedestrian) Died_Female',
                      'Offenders (Driver/Pedestrian) Died_Transgender',
                      'Offenders (Driver/Pedestrian) Died_Total']].sum()
victim_deaths = df[['Victims Died_Male',
                    'Victims Died_Female', 
                    'Victims Died_Transgender',
                    'Victims Died_Total']].sum()
summary = pd.DataFrame({
    'Offenders': offender_deaths,
    'Victims': victim_deaths
}).transpose()
summary[['Offenders (Driver/Pedestrian) Died_Male',
         'Offenders (Driver/Pedestrian) Died_Female',
         'Offenders (Driver/Pedestrian) Died_Transgender']].plot(kind='bar', stacked=True)

plt.title('Deaths by Gender (Offenders)')
plt.ylabel('Number of Deaths')
plt.show()

summary[['Victims Died_Male',
         'Victims Died_Female',
         'Victims Died_Transgender']].plot(kind='bar', stacked=True)
plt.title('Deaths by Gender (Victims)')
plt.ylabel('Number of Deaths')
plt.show()


summary[['Offenders (Driver/Pedestrian) Died_Total',
         'Victims Died_Total']].plot(kind='bar', color=['blue', 'orange'])

plt.title('Total Deaths (Offenders vs Victims)')
plt.ylabel('Total Deaths')
plt.show()
```

```python
total_accidents_by_state = df.groupby('States/UTs').agg({
    'Offenders (Driver/Pedestrian) Died_Total': 'sum',
    'Victims Died_Total': 'sum',
    'Victims Died_Male':'sum',
    'Victims Died_Female':'sum',
    
}).reset_index()


total_accidents_by_state['Total_Accidents'] = total_accidents_by_state['Offenders (Driver/Pedestrian) Died_Total'] + total_accidents_by_state['Victims Died_Total']


total_accidents_by_state
```

```python
States_highest_No_accidents = total_accidents_by_state.loc[total_accidents_by_state['Total_Accidents'].idxmax()]

States_highest_No_accidents
```

```python
gender_deaths = df.groupby('States/UTs').agg({
    'Offenders (Driver/Pedestrian) Died_Total': 'sum',
    'Victims Died_Total': 'sum',
    'Victims Died_Male':'sum',
    'Victims Died_Female':'sum',
    'Victims Died_Transgender':'sum',
    'Offenders (Driver/Pedestrian) Died_Male':'sum',
    'Offenders (Driver/Pedestrian) Died_Female':'sum',
    'Offenders (Driver/Pedestrian) Died_Transgender':'sum',
    
}).reset_index()
```
